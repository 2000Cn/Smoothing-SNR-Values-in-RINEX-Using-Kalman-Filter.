# Reading an Obs Files 

```python
OBS_B = gr.load('C:/Users/chandan/Downloads/test20200617-2359.20O',tlim=['2020-06-18T17:59', '2020-06-18T18:13'],fast=False)
```

```python
gr.obsheader2('C:/Users/chandan/Downloads/test20200617-2359.20O')
```

```python
OBS_B
```

### Showing different SV

```python
OBS_B.sv 
```

### Reading Snr values For GPS satellites

```python
OBS_B['S1'].sel(sv='G25').isnull()
```

### Filling the null values with zero

```python
clean_OBS_G=np.array([ 0 if np.isnan(ele) else ele for ele in OBS_B['S1'].sel(sv='G25') ])
```

```python
clean_OBS_G
```

# Plotting the curve For G25

```python
sns.set_style('darkgrid')
plt.figure(figsize = (10, 4))
plt.plot(clean_OBS_G,c='black')
```

# Now smoothing the curve by applying kalman Filter

```python
kf_SNR = KalmanFilter(transition_matrices = [1],
                  observation_matrices = [1],
                  initial_state_mean = clean_OBS_G[0],
                  initial_state_covariance = 1,
                  observation_covariance=1,
                  transition_covariance=.01)

state_means_SNR,_ = kf_SNR.filter(clean_OBS_G)
state_means_SNR = state_means_SNR.flatten()



fig, axis = plt.subplots(1,2, figsize = (20, 5))
axis[0].plot(clean_OBS_G,c='black')
axis[0].set_title("Before  Smoothing")
axis[1].plot(state_means_SNR,c='black')
axis[1].set_title("After Smoothing")
plt.show()
```

```python
state_means_SNR
```

```python
OBS_B['S1'].sel(sv='G25').values
```

```python
df = pd.DataFrame({'rough_data' : OBS_B['S1'].sel(sv='G25'), 'smooth_data' :state_means_SNR})
```

```python
df
```

```python
df.to_csv('smooth data_sv=G25.csv',index= False)
```
